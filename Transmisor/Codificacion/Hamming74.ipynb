```python
import time
import pandas as pd

def codificar_hamming(datos):
    datos_codificados = []

    # Dividir los datos en grupos de 4 bits y aplicar Hamming (7,4) a cada grupo
    for i in range(0, 8, 4):
        segmento_datos = datos[i:i+4]
        # Calcular los bits de paridad (P1, P2, P3)
        p1 = segmento_datos[0] ^ segmento_datos[1] ^ segmento_datos[3]
        p2 = segmento_datos[0] ^ segmento_datos[2] ^ segmento_datos[3]
        p3 = segmento_datos[1] ^ segmento_datos[2] ^ segmento_datos[3]
        # Combinar datos y bits de paridad
        segmento_codificado = [p1, p2, segmento_datos[0], p3, segmento_datos[1], segmento_datos[2], segmento_datos[3]]
        datos_codificados.extend(segmento_codificado)

    return datos_codificados

def decodificar_hamming(datos_codificados):
    datos_decodificados = []

    # Dividir los datos codificados en grupos de 7 bits y decodificar cada grupo
    for i in range(0, len(datos_codificados), 7):
        segmento_codificado = datos_codificados[i:i+7]
        # Calcular bits de paridad
        p1 = segmento_codificado[0]
        p2 = segmento_codificado[1]
        p3 = segmento_codificado[3]
        # Calcular errores
        e1 = p1 ^ segmento_codificado[2] ^ segmento_codificado[4] ^ segmento_codificado[6]
        e2 = p2 ^ segmento_codificado[2] ^ segmento_codificado[5] ^ segmento_codificado[6]
        e3 = p3 ^ segmento_codificado[4] ^ segmento_codificado[5] ^ segmento_codificado[6]
        # Calcular la posición del error
        posicion_error = e1 + e2 * 2 + e3 * 4
        # Corregir el error
        if posicion_error != 0:
            segmento_codificado[posicion_error - 1] ^= 1
        # Extraer datos
        datos_decodificados.extend([segmento_codificado[2], segmento_codificado[4], segmento_codificado[5], segmento_codificado[6]])

    return datos_decodificados

# Definir la función para medir el tiempo de ejecución
def medir_tiempo(funcion_script, num_ejecuciones=200):
    tiempos = []
    for _ in range(num_ejecuciones):
        inicio = time.time()
        funcion_script()
        fin = time.time()
        tiempo_ejecucion = fin - inicio
        tiempos.append(tiempo_ejecucion)
    tiempo_promedio = sum(tiempos) / num_ejecuciones
    return tiempo_promedio

# Define aquí la función o script que deseas medir
def ejecutar_script():
    archivo_excel = "C:\\Users\\jeiso\\Documents\\Jupyter NoteBook\\Datos\\Dataset.xlsx"
    datos_excel = pd.read_excel(archivo_excel)
    datos_excel['BPM'] = datos_excel['BPM'].round().astype(int)
    lista_datos = datos_excel['BPM'].apply(lambda x: [int(bit) for bit in format(x, '08b')]).tolist()
    lista_datos_codificados = []
    lista_datos_decodificados = []
    for datos in lista_datos:
        datos_codificados = codificar_hamming(datos)
        datos_decodificados = decodificar_hamming(datos_codificados)
        lista_datos_codificados.append(datos_codificados)
        lista_datos_decodificados.append(datos_decodificados)
    
    # Imprimir los datos codificados y decodificados
    print("Datos Codificados:")
    for datos_codificados in lista_datos_codificados:
        print(datos_codificados)
    
    print("\nDatos Decodificados:")
    for datos_decodificados in lista_datos_decodificados:
        print(datos_decodificados)

# Medir el tiempo de ejecución del script
tiempo_promedio_ejecucion = medir_tiempo(ejecutar_script)
print("Tiempo promedio de ejecución:", tiempo_promedio_ejecucion, "s")
```
